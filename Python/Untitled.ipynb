```python
from openpyxl import load_workbook
import pylab as pylab
import numpy as np
from matplotlib.mlab import griddata
import matplotlib.pyplot as plt
```

```python
def graficar(matrix,rango,titulo,folder):
    fig, ax = plt.subplots()
    fig.set_size_inches(8.5,7)
    fig.
    im=ax.imshow(matrix, aspect='auto', interpolation='none', origin='lower')
    #print ax.get_xticks()
    stepString=''
    if rango==41:
        ran=(rango-1)/4
        x=np.linspace(0,rango-1,ran-1)
        labels=np.linspace(-2,2,ran-1)
        stepString='(step=0.1mm)'
    else:
        ran=(rango-1)/4
        x=np.linspace(0,rango-1,ran)
        labels=np.linspace(-2,2,ran)
        stepString='(step=0.2mm)'
    
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.set_yticks(x)
    ax.set_yticklabels(labels)
    ax.set_xlabel('y')
    ax.set_ylabel('x')
    ax.set_xlim(0,rango-1)
    ax.set_ylim(0,rango-1)
    ax.set_title(titulo)
    labelColorBar=''+stepString
    cmap=plt.colorbar(im,ax=ax, cmap='jet',label=labelColorBar)
    
    fig.savefig(folder+titulo+'.png')
    plt.show()
```

```python
y=np.linspace(-4,4)
x=-y

plt.figure(figsize=(10,10))
plt.plot(y,y)
plt.savefig('idealPositiveCorrelation.png')
plt.show()


plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.savefig('idealNegativeCorrelation.png')
plt.show()

```
